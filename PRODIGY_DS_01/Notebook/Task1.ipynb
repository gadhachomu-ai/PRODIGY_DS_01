## Data Cleaning and Transformation

```python
# Drop the '2025' and 'Unnamed: 70' columns as they are completely empty
df_cleaned = df.drop(columns=['2025', 'Unnamed: 70'])

# Drop 'Indicator Name' and 'Indicator Code' as they appear to be constant and redundant for this dataset
df_cleaned = df_cleaned.drop(columns=['Indicator Name', 'Indicator Code'])

print("Dropped '2025', 'Unnamed: 70', 'Indicator Name', and 'Indicator Code' columns.")

display(df_cleaned.head())
```

```python
# Melt the DataFrame to transform year columns into 'Year' and 'Population' columns
# Identify columns that represent years (population data)
year_cols = [col for col in df_cleaned.columns if col.isdigit()]

# Identify ID variables
id_vars = ['Country Name', 'Country Code']

df_long = df_cleaned.melt(id_vars=id_vars, value_vars=year_cols, var_name='Year', value_name='Population')

# Convert 'Year' to integer type
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce').astype(int)

# Convert 'Population' to numeric type, coercing errors to NaN
df_long['Population'] = pd.to_numeric(df_long['Population'], errors='coerce')

print("DataFrame transformed to long format.")
display(df_long.head())
df_long.info()
```

## Handling Missing Values

```python
# Check for missing values in the transformed DataFrame
missing_values = df_long.isnull().sum()
print("Missing values after transformation:")
print(missing_values[missing_values > 0])

# Given that population data is inherently sequential, forward fill (ffill) and then backward fill (bfill)
# within each country group is a reasonable approach for handling missing population data.
# This ensures that gaps are filled with the last known population value, and then any remaining
# gaps at the beginning are filled with the first known value.

df_filled = df_long.copy()

df_filled['Population'] = df_filled.groupby('Country Code')['Population'].transform(lambda x: x.ffill().bfill())

print("\nMissing values after forward and backward fill:")
print(df_filled.isnull().sum()[df_filled.isnull().sum() > 0])

display(df_filled.head())
```

## Filtering out Aggregate Entities

```python
# Based on common patterns in World Bank data and observation from previous plots, filter out aggregate entities.
# These often have 'Country Code' values that are not standard 3-letter country codes or are descriptive names.
# A common approach is to exclude codes that are not ISO Alpha-3 codes or are obviously region aggregates.
# For this dataset, we can identify and remove rows where 'Country Code' corresponds to aggregates.

# List of common aggregate country codes/names to exclude (can be expanded if needed)
# This list is derived from observing the 'top_10_countries' output which included non-country entities.
aggregate_codes = [
    'WLD', 'IBT', 'LMY', 'MIC', 'IBD', 'EMU', 'LMC', 'UMC', 'EAS', 'LTE',
    'ECS', 'EUU', 'HIC', 'PRP', 'OEC', 'SSF', 'LAC', 'MNA', 'NAC', 'SAS',
    'SSA', 'SST', 'SXY', 'XZN', 'XKX', 'TLA', 'TSA', 'TEA', 'TEC', 'TSS'
]

df_countries = df_filled[~df_filled['Country Code'].isin(aggregate_codes)].copy()

print(f"Original number of entries: {len(df_filled)}")
print(f"Number of entries after filtering aggregates: {len(df_countries)}")

display(df_countries.head())
```

## Re-running EDA with Filtered Data

```python
# Re-run 'Population Distribution (Latest Year) - Histogram' with filtered data
plt.figure(figsize=(12, 7))
sns.histplot(df_countries[df_countries['Year'] == latest_year_filtered]['Population'].dropna(), bins=30, kde=True)
plt.title(f'Distribution of Population Across Countries in {latest_year_filtered} (Excluding Aggregates)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
